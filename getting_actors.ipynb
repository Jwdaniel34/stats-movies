```python
c.execute("""SELECT imdbid FROM Moovies.john mj
                JOIN Moovies.money mm
                ON mm.Title = mj.movieTitle AND mm.Released = mj.released""")
```

```python
df = pd.DataFrame(c.fetchall())
```

```python
for i in MIDS:
    req = requests.get(f'http://www.omdbapi.com/?i={i}&apikey=62416be1')
    try:
        final.append(req.json())
    except JSONDecodeError:
        pass
```

```python
from dateutil import parser
```

```python
for j in range(len(final)):
    try:
        final[j]['Released'] = parser.parse(final[j]['Released']).isoformat().strip('T00:00:00')
    except ValueError:
        continue
```

```python
for m in final:
    m['Actors'] = m['Actors'].split(', ')
```

```python
actors = []
```

```python
for movie in final:
    for actor in movie['Actors']:
        actors.append({'imdbid': movie['imdbID'],
                       'actor': actor})
```

```python
def add_to_actors_table(actors):
    # for each business that is in our results from our API request
    for i in range(len(actors)):
        # sql statement to add row
        add_actors = ("""INSERT INTO actors (imdbid, actor)
                                VALUES (%s, %s)
                        """)
        try:
            # note: location is made of a two-item list -- the first address line
            # and the city/state/etc.
            actor_values = (actors[i]['imdbid'],
                            actors[i]['actor'])

            c.execute("USE Moovies") #gets us into the yelp db instance
            c.execute(add_actors, actor_values)

        except mysql.connector.IntegrityError: # here to skip existing entries
            continue
        except KeyError: # here to skip the movies that are missing values
            continue

    cnx.commit()
```

```python
add_to_actors_table(actors)
```
