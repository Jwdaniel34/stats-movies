```python
i=0
dates_and_names = []
money_list= []
for i in range(1, 5800, 100):
    the_numbers_url = 'https://www.the-numbers.com/movie/budgets/all/{}'.format(i)
    r = requests.get(the_numbers_url)
    c = r.content
    soup = BeautifulSoup(c, 'html.parser')
    saved_data = scrapemovies(the_numbers_url)
    movie_budgets_gross = []
    for movie in saved_data:
        movie_budgets_gross.append(movie.text.strip())
    m=0
    while m<len(movie_budgets_gross):
        money_list.append(movie_budgets_gross[m:m+4])
        m+=4
    movies_list = []
    for movie in saved_data:
        movies_list.extend(movie.parent.select('a'))
    movie_names_dates = []
    for movie in movies_list:
        movie_names_dates.append(movie.text.strip())
    new_list_name_date=[]
    n = 0
    while n<len(movie_names_dates):
        new_list_name_date.append(movie_names_dates[n:n+2])
        n+=2
    x = 0
    for x in range(0, len(new_list_name_date), 4):
        dates_and_names.append(new_list_name_date[x])
```

```python
def add_tables_to_db(TABLES, cursor, db_name):
    from mysql.connector import errorcode
    for table_name in TABLES:
        table_description = TABLES[table_name]
        try:
            print("Creating table {}: ".format(table_name), end='')
            cursor.execute(f"""USE {db_name}""")
            cursor.execute(table_description)
        except mysql.connector.Error as err:
            if err.errno == errorcode.ER_TABLE_EXISTS_ERROR:
                print("already exists.")
            else:
                print(err.msg)
        else:
            print("OK")
```

```python
def add_to_moneytable(movie_lstofdicts):
    #for each movie in the list of dictionaries
    for i in range(len(movie_lstofdicts)):
        add_money = ("""INSERT INTO money VALUES (%s, %s, %s, %s, %s)""")
        try:
            money_values = (movie_lstofdicts[i]['Title'],
                            movie_lstofdicts[i]['Released'],
                            movie_lstofdicts[i]['Budget'],
                            movie_lstofdicts[i]['GrossDomestic'],
                            movie_lstofdicts[i]['GrossWorldWide'])
            
            cursor.execute("USE Moovies") #gets us into the Moovies db instance
            cursor.execute(add_money, money_values)
        
        except mysql.connector.IntegrityError: # here to skip existing entries
            continue
#         except KeyError: # here to skip the movies that are missing values, such as price
#             continue
        
            
            
    cnx.commit()
#     close_connection(cnx, cursor)
```
