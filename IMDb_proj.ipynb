```python
df = pd.read_csv('title.basics.tsv', sep="\t", dtype={'tconst': str, 
                                                      'titleType': str, 
                                                      'primaryTitle': str, 
                                                      'originalTitle': str,
                                                      'isAdult': int,
                                                      'startYear': object,
                                                      'endYear': object,
                                                      'runtimeMinutes': object,
                                                      'genres': str
                                                     })
```

```python
df.head()
```

```python
movie_ids = [x for x in list(df['tconst'])]
```

```python
movie_ids
```

```python
movie_years = [x for x in list(df['startYear'])]
```

```python
args = []
for i in range(len(movie_ids)):
    if movie_years[i] >= '2000':
        if movie_years[i] != '\\N':
            args.append((movie_ids[i], movie_years[i]))
```

```python
# [args[i] for i in range(len(args)) if args[i] >= '2000']
len(args)
```

```python
moobies = []
for i in range(1000):
    movie = args[i][0]
    req = requests.get(f'http://www.omdbapi.com/?i={movie}&apikey=30343807')
    moobies.append(req.json())
```
