```python
def add_to_paul_table(movies):
    # for each business that is in our results from our API request
    for i in range(len(movies)):
        # sql statement to add row
        add_movie = ("""INSERT INTO paul
                                VALUES (%s, %s, %s, %s)
                        """)
        try:
            # note: location is made of a two-item list -- the first address line
            # and the city/state/etc.
            movie_values = (movies[i]['Title'],
                            movies[i]['Released'],
                            movies[i]['imdbID'],
                            movies[i]['Director'])

            cursor.execute("USE Moovies") #gets us into the yelp db instance
            cursor.execute(add_movie, movie_values)

        except mysql.connector.IntegrityError: # here to skip existing entries
            continue
        except KeyError: # here to skip the movies that are missing values
            continue

    cnx.commit()
    
```

```python
add_to_paul_table(p)
```
