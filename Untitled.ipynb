```python
import matplotlib.pyplot as plt
```

```python
def plot_genre_ratings():
    import mysql.connector
    import matplotlib.pyplot as plt
    import config
    import pandas as pd
    
    cnx = mysql.connector.connect(
        host = config.host,
        user = config.user,
        passwd = config.password)

    cursor = cnx.cursor()
    
    cursor.execute("""SELECT gi.genreid, genreName, AVG(mj.rating) as AvgRating, COUNT(DISTINCT imdbid) as NumOfMovies
                    FROM Moovies.movie_genres mg
                    JOIN Moovies.genreid gi ON mg.genreId = gi.genreid
                    JOIN Moovies.john mj USING(imdbid)
                    GROUP BY gi.genreid """)
    
    df2 = pd.DataFrame(cursor.fetchall())
    df2.columns = [x[0] for x in cursor.description]
    
    counting = [x for x in df2['NumOfMovies']]
    ratings = [r for r in df2['AvgRating']]
    ratings_wavg = [(ratings[i]*10) * (counting[i]/sum(counting)) for i in range(len(ratings))]
    
    figs, ax1 = plt.subplots(figsize=(15,7))
    ax1.bar(df2['genreName'], ratings_wavg, color = 'blue')
    ax1.set_xlabel('Genres')
    ax1.set_ylabel('Rating')
    ax1.set_title('Genre VS Ratings')
    ax1.tick_params(labelrotation=90)

    plt.show()
```

```python
plot_genre_ratings()
```
