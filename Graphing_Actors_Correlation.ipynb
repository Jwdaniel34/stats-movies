```python
import config
import csv
from mysql.connector import errorcode
import pandas as pd
import mysql.connector as mysql
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
df = pd.read_csv("ActorsAVGprofitGen.csv")
toptencomdir_rating = df[0:15]
actor = toptencomdir_rating.actor
avg_rating = toptencomdir_rating.Wavgrating


fig, ax = plt.subplots(figsize = (10, 6))

ax.bar(actor, avg_rating)
ax.set_xticklabels(actor, rotation= 65)
ax.set_ylabel('Average Rating')
ax.set_title('Actors and Rating')
plt.show()
```
